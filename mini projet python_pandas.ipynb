```python
import pandas as pd

# Créer un DataFrame avec les données
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05',
             '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10'],
    'Produit': ['Clavier', 'Souris', 'Écran', 'Chaise', 'Bureau', 
                'Clavier', 'Souris', 'Bureau', 'Écran', 'Chaise'],
    'Catégorie': ['Informatique', 'Informatique', 'Informatique', 'Bureau', 'Bureau', 
                  'Informatique', 'Informatique', 'Bureau', 'Informatique', 'Bureau'],
    'Quantité': [5, 10, 2, 3, 1, 7, 4, 2, 1, 5],
    'Prix Unitaire (€)': [29.99, 19.99, 149.99, 89.90, 199.99, 
                          29.99, 19.99, 199.99, 149.99, 89.90],
    'Vendeur': ['Alice', 'Bob', 'Claire', 'David', 'Eva', 
                'Alice', 'Bob', 'Eva', 'Claire', 'David'],
    'Région': ['Île-de-France', 'Occitanie', 'Auvergne-Rhône-Alpes', 'Nouvelle-Aquitaine', 'Bretagne',
               'Île-de-France', 'Occitanie', 'Bretagne', 'Île-de-France', 'Auvergne-Rhône-Alpes']
}

df = pd.DataFrame(data)

# Afficher les 5 premières lignes pour vérifier le DataFrame
df.head()
```

```python
# Sauvegarder le DataFrame en fichier CSV
df.to_csv('ventes_produits.csv', index=False)

# Afficher un message pour confirmer que le fichier a été créé
print("Le fichier CSV a été créé avec succès.")
```

```python
# Vérification des informations générales
df.info()
```

```python
# Vérification des valeurs manquantes
df.isnull().sum()
```

```python
#Étape 3 : Manipulation des données (adapté à ton fichier)

# Sélectionner les colonnes 'Produit', 'Prix Unitaire (€)', et 'Quantité'
colonnes_seleccionnees = df[['Produit', 'Prix Unitaire (€)', 'Quantité']]

# Afficher le DataFrame avec ces colonnes
print(colonnes_seleccionnees)
```

```python
# 2. Filtrer les Vendeur avec une Prix Unitaire (€) > 30.0 (ajuste le seuil selon ton critère)
df_filtré = df[df['Prix Unitaire (€)'] > 30.0]

# Afficher les produits filtrés
print(df_filtré)
```

```python
# 3. Créer une nouvelle colonne 'Total Vente (€)' = Quantité * Prix Unitaire (€)
df['Total Vente (€)'] = df['Quantité'] * df['Prix Unitaire (€)']

# Afficher les premières lignes pour vérifier la nouvelle colonne
print(df[['Produit', 'Quantité', 'Prix Unitaire (€)', 'Total Vente (€)']].head())
```

```python
# Trier les données par 'Prix Unitaire (€)' de manière décroissante
df_sorted = df.sort_values(by='Prix Unitaire (€)', ascending=False)

# Afficher les données triées
print(df_sorted)
```

```python
# Statistiques descriptives
df.describe()
```

```python
#Analyse statistique:
import pandas as pd
# Statistiques descriptives
print("Statistiques descriptives :")
print(df[['Prix Unitaire (€)', 'Quantité']].describe())

# Médiane
median_prix = df['Prix Unitaire (€)'].median()
median_quantité = df['Quantité'].median()
print(f"Médiane du prix unitaire : {median_prix}")
print(f"Médiane de la quantité : {median_quantité}")

# Écart-type
ecart_type_prix = df['Prix Unitaire (€)'].std()
ecart_type_quantité = df['Quantité'].std()
print(f"Écart-type du prix unitaire : {ecart_type_prix}")
print(f"Écart-type de la quantité : {ecart_type_quantité}")

# Grouper les données par Vendeur et calculer la somme des ventes
df['Total Vente (€)'] = df['Quantité'] * df['Prix Unitaire (€)']
df_grouped = df.groupby('Vendeur')['Total Vente (€)'].sum()
print("\nSomme des ventes par Vendeur :")
print(df_grouped)

# Grouper par Vendeur et calculer la moyenne des ventes
df_grouped_avg = df.groupby('Vendeur')['Total Vente (€)'].mean()
print("\nMoyenne des ventes par Vendeur :")
print(df_grouped_avg)
```

```python
#Visualisation des données:
import matplotlib.pyplot as plt

# Tracer un histogramme pour les prix unitaires
plt.figure(figsize=(8,6))
df['Prix Unitaire (€)'].hist(bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution des Prix Unitaires (€)')
plt.xlabel('Prix Unitaire (€)')
plt.ylabel('Fréquence')
plt.grid(True)
plt.show()
```

```python
# Calculer le total des ventes par date
df['Total Vente (€)'] = df['Quantité'] * df['Prix Unitaire (€)']
df['Date'] = pd.to_datetime(df['Date'])  # S'assurer que la colonne 'Date' est au bon format

# Tracer une courbe des ventes totales par date
plt.figure(figsize=(10,6))
plt.plot(df['Date'], df['Total Vente (€)'], marker='o', color='orange', linestyle='-', markersize=8)

plt.title('Ventes Totales au Fil du Temps')
plt.xlabel('Date')
plt.ylabel('Total Vente (€)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotation des dates pour une meilleure lisibilité
plt.show()
```

```python
# Tracer un boxplot pour la colonne 'Prix Unitaire (€)'
plt.figure(figsize=(8,6))
plt.boxplot(df['Prix Unitaire (€)'], vert=False, patch_artist=True, notch=True, 
            boxprops=dict(facecolor='skyblue', color='blue'),
            whiskerprops=dict(color='blue'),
            medianprops=dict(color='red'))

plt.title('Boxplot des Prix Unitaires (€)')
plt.xlabel('Prix Unitaire (€)')
plt.show()
```

```python
# Créer un graphique avec deux sous-graphes
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Histogramme sur le premier sous-graphique
axes[0].hist(df['Prix Unitaire (€)'], bins=10, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution des Prix Unitaires (€)')
axes[0].set_xlabel('Prix Unitaire (€)')
axes[0].set_ylabel('Fréquence')

# Courbe sur le second sous-graphique
axes[1].plot(df['Date'], df['Total Vente (€)'], marker='o', color='orange', linestyle='-', markersize=8)
axes[1].set_title('Ventes Totales au Fil du Temps')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Total Vente (€)')
axes[1].tick_params(axis='x', rotation=45)

# Afficher la figure combinée
plt.tight_layout()
plt.show()
```
